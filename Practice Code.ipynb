```python
int2str = {}

ascii_val = 48

for rem in range(0,20):
    
    if rem == 10:
        
        ascii_val = 65
        
    int2str[rem] = chr(ascii_val)
    
    ascii_val += 1
```

```python
int2str
```

```python
def dec2base20(number):
    
    print("The Number",number,"is prime, therefore evaluating it's Base-20 equivalent")
    
    base20 = ""
    
    number_copy = number
    
    while number_copy > 0:
        
        remainder = number_copy % 20
        
        base20 += int2str[remainder]
        
        number_copy = number_copy//20
        
    return base20[::-1]
```

```python
def dec2base16(number):
    
    print("The number",number,"is odd, therefore evaluating it's Base-16 equivalent")
    
    base16 = ""
    
    number_copy = number
    
    while number_copy > 0:
        
        remainder = number_copy % 16
        
        base16 += int2str[remainder]
        
        number_copy = number_copy//16
        
    return base16[::-1]
```

```python
rand_nos = np.random.randint(low=512,high=2048,size=1000)
```

```python
odd_base_16_eqs = []
prime_base_20_eqs = []
```

```python
for random_number in rand_nos:

    if random_number % 2 != 0:

        for divisor in range(2,(random_number//2)+1):

            if random_number % divisor == 0:

                break

        if divisor == (random_number//2):

            base20_eq = dec2base20(random_number)
            
            prime_base_20_eqs.append(base20_eq)

            print("The Base-20 equivalent of number",random_number,"is",base20_eq)
            
            print("\n\n")

        else:

            base16_eq = dec2base16(random_number)
            
            odd_base_16_eqs.append(base16_eq)

            print("The base-16 equivalent of number",random_number,"is",base16_eq)
            
            print("\n\n")

    else:

        print("Since the number",random_number,"is Even, therefore not converting it")
        
        print("\n\n")
```
